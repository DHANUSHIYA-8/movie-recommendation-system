```python
# Step 1: Install necessary packages
!pip install ipywidgets --quiet

# Step 2: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact

# Step 3: Load dataset
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
df = pd.read_csv(url)

# Step 4: Clean & reshape data
df = df.drop(['Lat', 'Long'], axis=1)
long_df = df.melt(id_vars=["Province/State", "Country/Region"], var_name="Date", value_name="Confirmed")
long_df['Date'] = pd.to_datetime(long_df['Date'])

# Step 5: Build interactive plot function
countries = long_df['Country/Region'].unique()

def plot_country(country_name):
    data = long_df[long_df['Country/Region'] == country_name]
    grouped = data.groupby("Date").sum().reset_index()
    plt.figure(figsize=(10,5))
    plt.plot(grouped['Date'], grouped['Confirmed'], marker='o', linestyle='-')
    plt.title(f'COVID-19 Confirmed Cases Over Time in {country_name}')
    plt.xlabel('Date')
    plt.ylabel('Confirmed Cases')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Step 6: Show interactive widget
interact(plot_country, country_name=sorted(countries));
```
